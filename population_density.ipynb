```python
df = pd.read_csv("/content/2021 부산 50_100_면적.csv",encoding="cp949")
df
```

```python
# 컬럼 확인

df.columns
```

```python
# 65세 ~ 85세,면적 추출
senior_in_busan = df[['행정구역','65~69세','70~74세','75~79세','80~84세','면적(㎢)']]
senior_in_busan
```

```python
# 1열 인덱스로 변경

senior_in_busan =senior_in_busan.set_index(keys='행정구역')
senior_in_busan
```

```python
# 타입확인
senior_in_busan.dtypes
```

```python
# 기준 : 동 당 부산 노인인구 수 = 부산 65~84세 인구수 / 부산 동 개수 

sum=0;
num_of_people = senior_in_busan[['65~69세','70~74세','75~79세','80~84세']].sum()

standard= num_of_people.sum()/len(senior_in_busan)
standard
```

```python
# 행정동별 연령구간 인구수 구하기

section=list()
col = ['65~69세','70~74세','75~79세','80~84세']
for i in senior_in_busan.index:
  sum=0
  for j in col:
    sum= sum+ senior_in_busan.loc[i,j]
  section.append(sum)

# section

# senior_in_busan에 연령구간 인구수 열 추가

senior_in_busan['연령구간 인구수']=section

senior_in_busan
```

```python
# 동 당 부산 노인인구 수보다 많은 동만 추출

population={}  
sum=0

for i in senior_in_busan.index: 
  if senior_in_busan.loc[i,'연령구간 인구수']>standard: 
    population[i]=senior_in_busan.loc[i,'연령구간 인구수']

# population

len(population)
new_population = pd.DataFrame(population,index=["연령구간 인구수"])
new_population
population_value = new_population.transpose() # 행,열 변환
population_value
 
```

```python
## 동 당 노인 인구밀도 = 동 당 부산 노인인구 수 / 면적
density={}

for i in population_value.index:
  den = population_value.loc[i,'연령구간 인구수']/senior_in_busan.loc[i,'면적(㎢)']
  density[i]= den


density
population_density = pd.DataFrame(density,index=["인구밀도"])
# population_density

density_value =population_density.transpose() # 행,열 변환
density_value = density_value.sort_values(by=density_value.columns[0],ascending=False) # sort

density_value.head()
```
